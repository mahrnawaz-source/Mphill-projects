```python
# Core libraries
import numpy as np
import matplotlib.pyplot as plt

# TensorFlow / Keras
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical

# Evaluation metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
```

```python
# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize pixel values
X_train = X_train / 255.0
X_test = X_test / 255.0

# Save original labels for metrics
y_test_original = y_test.copy()

# One-hot encode labels
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)
```

```python
# Flatten images for MLP
X_train_mlp = X_train.reshape(-1, 28*28)
X_test_mlp = X_test.reshape(-1, 28*28)

mlp_model = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

mlp_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

mlp_model.summary()
```

```python
history_mlp = mlp_model.fit(
    X_train_mlp,
    y_train_cat,
    epochs=10,
    batch_size=128,
    validation_split=0.1
)
```

```python
# Predictions
y_pred_mlp = mlp_model.predict(X_test_mlp)
y_pred_mlp_classes = np.argmax(y_pred_mlp, axis=1)

# Metrics
mlp_accuracy = accuracy_score(y_test_original, y_pred_mlp_classes)
mlp_precision = precision_score(y_test_original, y_pred_mlp_classes, average='weighted')
mlp_recall = recall_score(y_test_original, y_pred_mlp_classes, average='weighted')

print("MLP Accuracy :", mlp_accuracy)
print("MLP Precision:", mlp_precision)
print("MLP Recall   :", mlp_recall)

print("\nClassification Report (MLP):\n")
print(classification_report(y_test_original, y_pred_mlp_classes))
```

```python
# Reshape for CNN (samples, height, width, channels)
X_train_cnn = X_train.reshape(-1, 28, 28, 1)
X_test_cnn = X_test.reshape(-1, 28, 28, 1)
```

```python
cnn_model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

cnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

cnn_model.summary()
```

```python
history_cnn = cnn_model.fit(
    X_train_cnn,
    y_train_cat,
    epochs=10,
    batch_size=128,
    validation_split=0.1
)
```

```python
# Predictions
y_pred_cnn = cnn_model.predict(X_test_cnn)
y_pred_cnn_classes = np.argmax(y_pred_cnn, axis=1)

# Metrics
cnn_accuracy = accuracy_score(y_test_original, y_pred_cnn_classes)
cnn_precision = precision_score(y_test_original, y_pred_cnn_classes, average='weighted')
cnn_recall = recall_score(y_test_original, y_pred_cnn_classes, average='weighted')

print("CNN Accuracy :", cnn_accuracy)
print("CNN Precision:", cnn_precision)
print("CNN Recall   :", cnn_recall)

print("\nClassification Report (CNN):\n")
print(classification_report(y_test_original, y_pred_cnn_classes))
```

```python
print("MODEL COMPARISON")
print("----------------------------")
print(f"MLP Accuracy  : {mlp_accuracy:.4f}")
print(f"CNN Accuracy  : {cnn_accuracy:.4f}")
print(f"MLP Precision : {mlp_precision:.4f}")
print(f"CNN Precision : {cnn_precision:.4f}")
print(f"MLP Recall    : {mlp_recall:.4f}")
print(f"CNN Recall    : {cnn_recall:.4f}")
```
